```python
from sklearn.cluster import KMeans
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("Happiness Questionnaire.csv")
df



```

```python
f1 = df['Score'].values
# generate another list
f2 = list(range(0, len(f1)))
X = np.array(list(zip(f1, f2)))
```

```python
kmeans = KMeans(n_clusters=3)
df['label'] = kmeans.fit_predict(df[['Score']])
df
```

```python

ax = df[df['label']==0].plot.scatter(x='Score', y='label', s=50, color='white', edgecolor='black')
df[df['label']==1].plot.scatter(x='Score', y='label', s=50, color='white', ax=ax, edgecolor='red')
df[df['label']==2].plot.scatter(x='Score', y='label', s=50, color='white', ax=ax, edgecolor='blue')

plt.scatter(kmeans.cluster_centers_.ravel(), [0.5]*len(kmeans.cluster_centers_), s=100, color='green', marker='*')
```
